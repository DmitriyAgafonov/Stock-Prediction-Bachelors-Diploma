### Fill NaN with interpolation and remove outliers

```python
df_ = df.copy()
```

```python
df_.columns
```

```python
def remove_outliers(df):
    Q1 = df.quantile(0.2)
    Q3 = df.quantile(0.8)
    IQR = Q3 - Q1
    return df[(df >= (Q1 - 1.5 * IQR)) & (df  <= (Q3 + 1.5 * IQR))]
```
